```python
# Function to remove a word and strip all words

def remove_and_strip(word_list, word_to_remove):
    # Strip spaces from all words
    stripped_list = [w.strip() for w in word_list]
    # Remove the given word if it exists
    result_list = [w for w in stripped_list if w != word_to_remove]
    return result_list

# Example usage
words = [" apple ", " banana", "cherry ", "banana "]
word_to_remove = "banana"

new_list = remove_and_strip(words, word_to_remove)
print(new_list)
```

```python
# add a static method in problem 2 to greet the user with hello

import math

class Calculator:
    def __init__(self, number):
        self.number = number

    def square(self):
        return self.number ** 2

    def cube(self):
        return self.number ** 3

    def square_root(self):
        return math.sqrt(self.number)

    @staticmethod
    def greet():
        print("Hello! Welcome to the Calculator.")


# Example usage
Calculator.greet()  # Call the static method without creating an object

num = float(input("Enter a number: "))
calc = Calculator(num)

print(f"Square of {num} is {calc.square()}")
print(f"Cube of {num} is {calc.cube()}")
print(f"Square root of {num} is {calc.square_root()}")
```

```python
# write a class Train which has methods to book a ticket, get status (no of seats) and get fare information of trains running under UK railways.

class Train:
    def __init__(self, name, total_seats, fare):
        self.name = name
        self.total_seats = total_seats  # Total seats in the train
        self.booked_seats = 0           # Initially no seats are booked
        self.fare = fare                # Fare per ticket

    def book_ticket(self, number_of_tickets=1):
        """Books tickets if seats are available"""
        if number_of_tickets <= 0:
            print("‚ùå Number of tickets should be at least 1.")
            return
        if self.booked_seats + number_of_tickets <= self.total_seats:
            self.booked_seats += number_of_tickets
            print(f"‚úÖ {number_of_tickets} ticket(s) booked successfully for {self.name}.")
        else:
            available = self.total_seats - self.booked_seats
            print(f"‚ùå Only {available} seat(s) available. Cannot book {number_of_tickets} tickets.")

    def get_status(self):
        """Returns the number of available seats"""
        available_seats = self.total_seats - self.booked_seats
        print(f"Train: {self.name}")
        print(f"Available seats: {available_seats}/{self.total_seats}")

    def get_fare_info(self):
        """Displays the fare per ticket"""
        print(f"Train: {self.name}")
        print(f"Fare per ticket: ¬£{self.fare}")


# Example usage
train1 = Train("London Express", 100, 50)
train2 = Train("Manchester Flyer", 80, 40)

# Book tickets
train1.book_ticket(3)
train2.book_ticket(5)

# Get seat status
train1.get_status()
train2.get_status()

# Get fare information
train1.get_fare_info()
train2.get_fare_info()
```

```python
# create a class C-2d vector and use it to create another class representing a 3-d vector 

class C2DVector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def display(self):
        print(f"2D Vector: ({self.x}, {self.y})")


class C3DVector(C2DVector):
    def __init__(self, x, y, z):
        super().__init__(x, y)
        self.z = z

    def display(self):
        print(f"3D Vector: ({self.x}, {self.y}, {self.z})")


# Example usage
v2d = C2DVector(3, 4)
v2d.display()   # Output: 2D Vector: (3, 4)

v3d = C3DVector(1, 2, 3)
v3d.display()   # Output: 3D Vector: (1, 2, 3)
```

```python
# create  a class pets from a class  Animals and further create class Dog from Pets . Add a method bark to class Dog

class Animals:
    def __init__(self):
        print("This is the Animal class.")

class Pets(Animals):
    def __init__(self):
        super().__init__()   # Calls Animals constructor
        print("This is the Pets class.")

class Dog(Pets):
    def __init__(self):
        super().__init__()   # Calls Pets constructor
        print("This is the Dog class.")

    def bark(self):
        print("Woof! Woof!")


# Example usage
d = Dog()        # Creates a Dog object
d.bark()         # Calls bark method
```

```python
# create a class Employee and Salary and increment properties to it . arite a method SalaryAfterIncrement method with a @property decorator with a setter changes the value of increment based on the salary

class Employee:
    def __init__(self, salary, increment):
        self.salary = salary
        self.increment = increment

    @property
    def SalaryAfterIncrement(self):
        return self.salary + (self.salary * self.increment / 100)

    @SalaryAfterIncrement.setter
    def SalaryAfterIncrement(self, new_salary):
        self.increment = ((new_salary - self.salary) / self.salary) * 100


# Example usage
e = Employee(50000, 10)
print("Initial salary:", e.salary)
print("Increment percentage:", e.increment)
print("Salary after increment:", e.SalaryAfterIncrement)

# Change salary after increment using setter
e.SalaryAfterIncrement = 60000  # updates increment based on new salary
print("\nAfter updating SalaryAfterIncrement to 60000:")
print("New increment percentage:", e.increment)
print("Updated SalaryAfterIncrement:", e.SalaryAfterIncrement)
```

```python
# write a class Complex to represent Complex numbers, along with overloaded operators + and * which adds and multiplies them

class Complex:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        # (a + bi) + (c + di) = (a + c) + (b + d)i
        return Complex(self.real + other.real, self.imag + other.imag)

    def __mul__(self, other):
        # (a + bi) * (c + di) = (ac - bd) + (ad + bc)i
        real_part = self.real * other.real - self.imag * other.imag
        imag_part = self.real * other.imag + self.imag * other.real
        return Complex(real_part, imag_part)

    def __str__(self):
        # Nicely formats the output when printed
        return f"{self.real} + {self.imag}i"


# Example usage
c1 = Complex(3, 2)
c2 = Complex(1, 7)

print("First Complex Number:", c1)
print("Second Complex Number:", c2)

print("\nAddition:", c1 + c2)
print("Multiplication:", c1 * c2)
```

```python
# write a class Vector representing a vector of n dimension . Overload the + and * operator which calculates the sum and the dot product of them

class Vector:
    def __init__(self, vec):
        self.vec = vec  # store vector as a list of numbers

    def __add__(self, other):
        # Element-wise addition of two vectors
        new_vec = [self.vec[i] + other.vec[i] for i in range(len(self.vec))]
        return Vector(new_vec)

    def __mul__(self, other):
        # Dot product of two vectors
        dot_product = sum(self.vec[i] * other.vec[i] for i in range(len(self.vec)))
        return dot_product

    def __str__(self):
        # Pretty print the vector
        return f"Vector({', '.join(map(str, self.vec))})"


# Example usage
v1 = Vector([1, 2, 3])
v2 = Vector([4, 5, 6])

print("First vector:", v1)
print("Second vector:", v2)

print("\nVector Addition:", v1 + v2)     # Vector addition
print("Dot Product:", v1 * v2)          # Dot product
```

```python
# override the __len__() method on Vector of problem 5 to display the dimension of the vector

class Vector:
    def __init__(self, vec):
        self.vec = vec  # list of vector components

    def __add__(self, other):
        new_vec = [self.vec[i] + other.vec[i] for i in range(len(self.vec))]
        return Vector(new_vec)

    def __mul__(self, other):
        dot_product = sum(self.vec[i] * other.vec[i] for i in range(len(self.vec)))
        return dot_product

    def __str__(self):
        # For a 3D vector, print as "7i + 8j + 10k"
        labels = ['i', 'j', 'k']
        return " + ".join(f"{self.vec[i]}{labels[i]}" for i in range(len(self.vec)))

    def __len__(self):
        # Return the dimension of the vector
        return len(self.vec)


# Example usage
v1 = Vector([7, 8, 10])
v2 = Vector([1, 2, 3])

print("First vector:", v1)
print("Second vector:", v2)
print("\nVector Addition:", v1 + v2)
print("Dot Product:", v1 * v2)
print("Dimension of v1:", len(v1))
```

```python
# wap to find the maximum of the numbers in a list using the reduce function

from functools import reduce

# Example list
numbers = [10, 23, 45, 12, 55, 67, 100, 3]

# Use reduce to find the maximum
max_number = reduce(lambda a, b: a if a > b else b, numbers)

print("Maximum number in the list:", max_number)
```
